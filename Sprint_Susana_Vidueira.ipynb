#2.
Diseñe e implemente una clase Inventario que utilice una estructura de datos (por ejemplo, un diccionario) para almacenar objetos de la clase Producto. La clase Inventario debe incluir métodos para realizar las siguientes operaciones:
• Agregar un producto al inventario
• Eliminar un producto del inventario
• Actualizar la información de un producto en el inventario
• Buscar un producto en el inventario por su id
• Listar todos los productos en el inventario, incluyendo información como la cantidad total de productos y el valor total del inventario

```python
class Producto:
    def __init__(self, id, nombre, descripcion, cantidad, precio):
        self.id = id
        self.nombre = nombre
        self.descripcion = descripcion
        self.cantidad = cantidad
        self.precio = precio

    def __str__(self):
        return f"ID: {self.id}, Nombre: {self.nombre}, Descripción: {self.descripcion}, Cantidad: {self.cantidad}, Precio: {self.precio}"

    def actualizar(self, nombre=None, descripcion=None, cantidad=None, precio=None):
        if nombre:
            self.nombre = nombre
        if descripcion:
            self.descripcion = descripcion
        if cantidad:
            self.cantidad = cantidad
        if precio:
            self.precio = precio

class Inventario:
    def __init__(self):
        self.productos = []

    def agregar_producto(self, producto):
        self.productos.append(producto)

    def eliminar_producto(self, id):
        for p in self.productos:
            if p.id == id:
                self.productos.remove(p)
                return True
        return False

    def buscar_producto(self, id):
        for p in self.productos:
            if p.id == id:
                return p
        return None

    def generar_reporte(self):
        if not self.productos:
            print("El inventario está vacío")
            return

        print("Reporte de Inventario:")
        for p in self.productos:
            print(p)

    def listar_productos(self):
        if not self.productos:
            print("El inventario está vacío")
            return

        print("Lista de Productos en el Inventario:")
        for p in self.productos:
            print(p)

        cantidad_total = sum(p.cantidad for p in self.productos)
        valor_total = sum(p.cantidad * p.precio for p in self.productos)
        print(f"Cantidad total de productos: {cantidad_total}")
        print(f"Valor total del inventario: {valor_total}")

# Ejemplo de uso
if __name__ == "__main__":
    inventario = Inventario()

    producto1 = Producto(1, "Jabon", "Jabon Liquido", 100, 500)
    producto2 = Producto(2, "Shampoo", "Shampoo en base a hierbas", 50, 1000)

    inventario.agregar_producto(producto1)
    inventario.agregar_producto(producto2)

    inventario.generar_reporte()

    producto_encontrado = inventario.buscar_producto(1)
    if producto_encontrado:
        producto_encontrado.actualizar(cantidad=90)

    inventario.generar_reporte()

    inventario.eliminar_producto(2)

    inventario.generar_reporte()

    inventario.listar_productos()
```
